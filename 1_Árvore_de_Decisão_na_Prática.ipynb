```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel(r"/content/drive/MyDrive/Árvore de Decisão/Banco de Dados/golf_df.xlsx")
```

```python
dados.head()
```

```python
# EXPLICATIVAS: clima, temp, umidade, vento
# TARGET: jogar

x = pd.get_dummies(dados[["clima", "temp", "umidade", "vento"]])
x.head()
```

```python
y = dados[["jogar"]]
y
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
modelo = arvore.fit(x, y)
```

```python
modelo
```

```python
fig = plt.figure(figsize=(70,30))

fig = tree.plot_tree(modelo, feature_names = x.columns, class_names = ["yes", "no"], filled = True)
```
