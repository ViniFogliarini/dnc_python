## Setup

```python
import pandas as pd
import numpy as np
```

```python
house = pd.read_excel(r'/content/drive/MyDrive/Colab Notebooks/Regressão Não Linear/house.xlsx')
house.head()
```

## Conhecendo e Tratando os Dados

```python
house.info()
```

```python
house = house.dropna()
```

```python
# One-hot encode

house = pd.get_dummies(house)
house.head()
```

```python
house.info()
```

## Separando as Variaveis

```python
# Labels are the values we want to predict
labels = np.array(house['SalePrice'])

# Remove the labels
# axis 1 refers to the columns
features = house.drop('SalePrice', axis = 1) # não alterar nome da variavel

# Saving feature names for later use
features_list = list(house.columns) # não alterar nome da variavel

# Convert to numpy array
house = np.array(features)
```

## Treino e Teste

```python
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn import metrics
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

## Ja sabemos que a melhor é GradientBoostingRegressor

```python
grb = GradientBoostingRegressor(n_estimators=1000)
grb.fit(train_features, train_labels)
gbr_pred = grb.predict(test_features)
```

```python
# Calculate the absolute errors
errors_gbr = abs(gbr_pred - test_labels)
r_sq = grb.score(features, labels)

# Print out the mean absolute error (mae)
print('Coeficiente de Determinação (R²):', r_sq)
print('MAE:', metrics.mean_absolute_error(test_labels, gbr_pred))
print('MSE:', metrics.mean_squared_error(test_labels, gbr_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test_labels, gbr_pred)))
```
