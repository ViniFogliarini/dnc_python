### Setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_lucro = pd.read_excel(r"/content/drive/MyDrive/Colab Notebooks/Limite_Credito_Dummy.xlsx")
df_lucro.head()
```

```python
# Correção no nome das colunas Gênero e Região
df_lucro.rename(columns={'Gênero': 'R'}, inplace=True)
df_lucro.rename(columns={'Região': 'Gênero'}, inplace=True)
df_lucro.rename(columns={'R': 'Região'}, inplace=True)
df_lucro.head()
```

### Transformando em Dummy

```python
df_lucro = pd.get_dummies(data=df_lucro, columns=['Escolaridade', 'Gênero', 'Região'], prefix=['Escolaridade', 'Gênero', 'Região'])
df_lucro.head()
```

```python
df_lucro.corr()
```

```python
sns.pairplot(df_lucro);
```

### Modelo OLS

```python
import statsmodels.formula.api as smf
```

```python
df_lucro.head()
```

```python
function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Gênero_F+Gênero_M+Região_RJ+Região_SP'
model = smf.ols(formula=function, data=df_lucro).fit()
print(model.summary())
```
